# Analyse des donnees

```python
import json
from collections import defaultdict
import spacy
from urllib.parse import urlparse
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Charger le modèle français de spaCy
try:
    nlp = spacy.load("fr_core_news_sm")
except:
    import subprocess
    subprocess.run(["python", "-m", "spacy", "download", "fr_core_news_sm"])
    nlp = spacy.load("fr_core_news_sm")

# Charger les données des catégories
def charger_donnees_categories(fichier_categories):
    """Charger les données des catégories depuis un fichier JSON."""
    with open(fichier_categories, "r", encoding="utf-8") as file:
        return json.load(file)

# Charger les données fusionnées et nettoyées
def charger_donnees_fusionnees(fichier_fusion):
    """Charger les données fusionnées et nettoyées depuis un fichier JSON."""
    with open(fichier_fusion, "r", encoding="utf-8") as file:
        return json.load(file)

# Extraire le nom de la catégorie depuis l'URL
def extraire_nom_categorie(url):
    """Extraire le nom de la catégorie depuis une URL."""
    return urlparse(url).path.strip("/").split("/")[-1].replace("-", " ").capitalize()

# Associer les produits aux catégories
def associer_produit_categorie(donnees_categories):
    """Associer chaque produit à une catégorie donnée."""
    produit_categorie = {}
    for categorie, produits in donnees_categories.items():
        for produit in produits:
            produit_categorie[produit["link"]] = categorie
    return produit_categorie

# Calculer les notes moyennes pondérées par catégorie
def calculer_notes_ponderees_par_categorie(donnees, produit_categorie):
    """Calculer les notes moyennes pondérées par catégorie en fonction des avis."""
    notes_par_categorie = defaultdict(lambda: {"total_notes_ponderees": 0, "total_avis": 0})
    for produit in donnees:
        lien = produit["link"]
        categorie = produit_categorie.get(lien, "Non catégorisé")
        note_moyenne = produit.get("average_rating")
        total_avis = produit.get("total_reviews")
        if note_moyenne is not None and total_avis is not None:
            notes_par_categorie[categorie]["total_notes_ponderees"] += note_moyenne * total_avis
            notes_par_categorie[categorie]["total_avis"] += total_avis
    notes_ponderees = {
        categorie: round(notes["total_notes_ponderees"] / notes["total_avis"], 2)
        if notes["total_avis"] > 0 else 0
        for categorie, notes in notes_par_categorie.items()
    }
    return notes_ponderees

# Générer un nuage de mots pour une catégorie
def generer_wordcloud(termes, categorie):
    """Générer un nuage de mots pour une catégorie donnée."""
    if termes:
        wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(dict(termes))
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.title(f"Nuage de mots - {categorie.capitalize()}")
        plt.savefig(f"{categorie}_wordcloud.png")
        plt.show()

# Générer un graphique des bigrammes
def visualiser_bigrammes(bigrammes, categorie):
    """Créer un graphique des bigrammes les plus fréquents pour une catégorie donnée."""
    if bigrammes:
        mots = [" ".join(bigramme[0]) for bigramme in bigrammes]
        frequences = [bigramme[1] for bigramme in bigrammes]
        plt.figure(figsize=(10, 5))
        plt.barh(mots, frequences, color="skyblue")
        plt.xlabel("Fréquence")
        plt.ylabel("Bigrammes")
        plt.title(f"Bigrammes les plus fréquents - {categorie.capitalize()}")
        plt.gca().invert_yaxis()
        plt.tight_layout()
        plt.savefig(f"{categorie}_bigrammes.png")
        plt.show()

# Visualiser les notes moyennes pondérées
def visualiser_notes_ponderees(notes_ponderees):
    """Créer un graphique à barres des notes moyennes pondérées par catégorie."""
    categories = [extraire_nom_categorie(categorie) for categorie in notes_ponderees.keys()]
    notes = list(notes_ponderees.values())
    plt.figure(figsize=(12, 6))
    plt.barh(categories, notes, color="skyblue")
    plt.xlabel("Note moyenne pondérée")
    plt.ylabel("Catégorie")
    plt.title("Notes moyennes pondérées par catégorie")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig("notes_ponderees_categorie.png")
    plt.show()

# Analyse des termes fréquents (unigrammes) par catégorie
def analyser_termes_frequents(donnees, produit_categorie, stop_words_personnalises=None):
    """Analyser les termes les plus fréquents (unigrammes) dans les commentaires par catégorie."""
    if stop_words_personnalises is None:
        stop_words_personnalises = set()
    commentaires_par_categorie = defaultdict(list)
    for produit in donnees:
        lien = produit["link"]
        categorie = produit_categorie.get(lien, "Non catégorisé")
        for commentaire in produit.get("comments", []):
            contenu_traite = commentaire.get("processed_content", "")
            if contenu_traite:
                doc = nlp(contenu_traite.lower())
                tokens = [
                    token.lemma_ for token in doc
                    if token.is_alpha and not token.is_stop and token.text not in stop_words_personnalises
                ]
                commentaires_par_categorie[categorie].extend(tokens)
    termes_par_categorie = {}
    for categorie, mots in commentaires_par_categorie.items():
        frequences = {mot: mots.count(mot) for mot in set(mots)}
        termes_par_categorie[categorie] = sorted(frequences.items(), key=lambda x: x[1], reverse=True)[:10]
    return termes_par_categorie

# Analyse des bigrammes fréquents par catégorie
def analyser_bigrammes_frequents(donnees, produit_categorie, stop_words_personnalises=None):
    """Analyser les bigrammes les plus fréquents dans les commentaires par catégorie."""
    if stop_words_personnalises is None:
        stop_words_personnalises = set()
    bigrammes_par_categorie = defaultdict(list)
    for produit in donnees:
        lien = produit["link"]
        categorie = produit_categorie.get(lien, "Non catégorisé")
        for commentaire in produit.get("comments", []):
            contenu_traite = commentaire.get("processed_content", "")
            if contenu_traite:
                doc = nlp(contenu_traite.lower())
                tokens = [
                    token.lemma_ for token in doc
                    if token.is_alpha and not token.is_stop and token.text not in stop_words_personnalises
                ]
                for i in range(len(tokens) - 1):
                    bigrammes_par_categorie[categorie].append((tokens[i], tokens[i + 1]))
    bigrammes_frequents = {}
    for categorie, bigrammes in bigrammes_par_categorie.items():
        frequences = defaultdict(int)
        for bigramme in bigrammes:
            frequences[bigramme] += 1
        bigrammes_frequents[categorie] = sorted(frequences.items(), key=lambda x: x[1], reverse=True)[:10]
    return bigrammes_frequents

# Ajouter une nouvelle couche de stop-words
nouveaux_stop_words = {"bon", "n", "l", "trop", "aimer", "xl", "shirt", "polo", "xxl", "adorer"}

# Chemins des fichiers
fichier_categories = r"C:\TOUS_MES_FICHIERS\ProjetPython\TravauxTechniques\1scraping\datasets\products\product_links.json"
fichier_fusion = r"C:\TOUS_MES_FICHIERS\ProjetPython\TravauxTechniques\2traitement\datasets\data_fusion_cleaned.json"

# Charger les données
print("Chargement des données...")
donnees_categories = charger_donnees_categories(fichier_categories)
donnees_fusionnees = charger_donnees_fusionnees(fichier_fusion)

# Associer produits aux catégories
produit_categorie = associer_produit_categorie(donnees_categories)

# Calcul des notes pondérées par catégorie
print("\nCalcul des notes moyennes pondérées par catégorie...")
notes_ponderees = calculer_notes_ponderees_par_categorie(donnees_fusionnees, produit_categorie)
print("Notes moyennes pondérées :")
for categorie, note in sorted(notes_ponderees.items(), key=lambda x: x[1], reverse=True):
    nom_categorie = extraire_nom_categorie(categorie)
    print(f"{nom_categorie}: {note}")

# Visualisation des notes pondérées
visualiser_notes_ponderees(notes_ponderees)

# Analyse des termes fréquents
print("\nAnalyse des termes les plus fréquents (unigrammes) par catégorie...")
termes_frequents = analyser_termes_frequents(donnees_fusionnees, produit_categorie, nouveaux_stop_words)
for categorie, termes in termes_frequents.items():
    nom_categorie = extraire_nom_categorie(categorie)
    print(f"{nom_categorie}: {termes}")
    generer_wordcloud(termes, nom_categorie)

# Analyse des bigrammes fréquents
print("\nAnalyse des bigrammes les plus fréquents par catégorie...")
bigrammes_frequents = analyser_bigrammes_frequents(donnees_fusionnees, produit_categorie, nouveaux_stop_words)
for categorie, bigrammes in bigrammes_frequents.items():
    nom_categorie = extraire_nom_categorie(categorie)
    print(f"{nom_categorie}: {bigrammes}")
    visualiser_bigrammes(bigrammes, nom_categorie)
```
